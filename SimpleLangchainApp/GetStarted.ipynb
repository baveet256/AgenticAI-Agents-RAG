```python
## scrape the data
from langchain_community.document_loaders import WebBaseLoader ## internally beautiful soup
```

```python
loader = WebBaseLoader("https://en.wikipedia.org/wiki/Wiki")
text = loader.load()
print(text) ## returns document here,  
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=100)
all_chunks = text_splitter.split_documents(text)
```

```python
## tovectors 
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
embeddings = OpenAIEmbeddings()
```

```python
data = [
    "The quick brown fox jumps over the lazy dog.",
    "Artificial intelligence is transforming technology.",
    "Python is a versatile programming language.",
    "FAISS helps in efficient similarity search.",
    "Embeddings convert text into vectors."
]
```

```python
import numpy as np
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.schema import Document
```

```python
# 3. Embed documents manually (list of vectors)
embedding_vectors = embeddings.embed_documents(data)

# 4. Convert to numpy array
embedding_vectors = np.array(embedding_vectors)

# 5. Create list of Document objects (required by FAISS.from_documents)
docs = [Document(page_content=text) for text in data]

# 6. Now create vectorstore with FAISS.from_documents (this should work)
vectorstore_db = FAISS.from_documents(docs, embeddings)
```

```python
vectorstore_db
```

```python
query = "How do i get better with AI?"
result = vectorstore_db.similarity_search(query)
result
```

```python
result[0].page_content
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain ### get a paragraph as a context
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    """
    Answer the following based on the provided context!!
    {input}
    <context>
    {context}
    </context>
    """
)

document_chain = create_stuff_documents_chain(llm,prompt)
document_chain
```

```python
document_chain.invoke({
    ## key value
    "input":""
    "context":""
})
## it works great but has no info of most relevant context. we just have to provide the context and give it to the llm
```

## Retrievers!!

✅ Execution Flow Now:

retrieval_chain.invoke({"input": "Should I learn Python for AI?"})

User sends:

{"input": "Should I learn Python for AI?"}
retrieval_chain sends "Should I learn Python for AI?" to the retriever.

The retriever finds relevant docs.

document_chain is called with:

{
  "input": "Should I learn Python for AI?",
  "context": retrieved_docs
}

The prompt is filled:

Answer the following based on the provided context:
Question: Should I learn Python for AI?

<context>
...retrieved docs here...
</context>

```python
retriever = vectorstore_db.as_retriever()
from langchain.chains import create_retrieval_chain
retriever_chain = create_retrieval_chain(retriever,document_chain)
response = retriever_chain.invoke({"input":"So should i start learning python to get started with AI?"})

# This defines a stuff document chain, meaning: 

# It takes all retrieved documents (as-is),

# Stuffs them into the {context} placeholder in your prompt,

# Then sends that prompt to the LLM to get a response.
```
